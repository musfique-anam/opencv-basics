```python
#Import Libraries

import cv2
import numpy as ap
import matplotlib.pyplot as plt
```

```python
#Read Image

img=cv2.imread('images/sample.jpg')

#Display Image

#Show using OpenCV(window)
cv2.imshow("Original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Show using matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Save the Image

cv2.imwrite('images/output.jpg',img)
```

```python
#Resize

resized=cv2.resize(img,(200,200))
plt.imshow(cv2.cvtColor(resized,cv2.COLOR_BGR2RGB))
plt.title("Resized Image(200x200)")
plt.axis("off")
plt.show()
```

```python
#Crop

cropped=img[800:2300,250:2000]
plt.imshow(cv2.cvtColor(cropped,cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
#Rotate

(h,w)=img.shape[:2] #ignores channels
center=(w/2,h/2) #rotate through middle of the image

M=cv2.getRotationMatrix2D(center,90,1.0)
rotated=cv2.warpAffine(img,M,(w,h))

plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("Rotated Image(90deg)")
plt.axis("off")
plt.show()
```

```python
#Flip

flip_horizontal=cv2.flip(img,1) #1=horizontal
flip_vertical=cv2.flip(img,0) #0=vertical

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB))
plt.title("Flip Horizontal")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB))
plt.title("Flip Vertical")
plt.axis("off")

plt.show()
```

```python
#Save transformed Images

cv2.imwrite('results/resized.jpg',resized)
cv2.imwrite('results/cropped.jpg',cropped)
cv2.imwrite('results/rotated.jpg',rotated)
cv2.imwrite('results/flip_horizontal.jpg',flip_horizontal)
cv2.imwrite('results/flip_vertical.jpg',flip_vertical)
```
